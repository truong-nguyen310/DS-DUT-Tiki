#### 2.2 Clean data 

```python
import pandas as pd

# Đọc dữ liệu từ tệp CSV
dt = pd.read_csv('clean_data/clean_data.csv')

# Tính toán giá trị trung bình của các cột 'stock_item_qty' và 'stock_item_max_sale_qty'
mean_stock_item_qty = dt['stock_item_qty'].mean()
mean_stock_item_max_sale_qty = dt['stock_item_max_sale_qty'].mean()

# Làm tròn giá trị trung bình
mean_stock_item_qty_rounded = round(mean_stock_item_qty)
mean_stock_item_max_sale_qty_rounded = round(mean_stock_item_max_sale_qty)

# Thay thế các giá trị NaN bằng giá trị trung bình đã được làm tròn
dt['stock_item_qty'] = dt['stock_item_qty'].fillna(mean_stock_item_qty_rounded)
dt['stock_item_max_sale_qty'] = dt['stock_item_max_sale_qty'].fillna(mean_stock_item_max_sale_qty_rounded)

# Thay thế các giá trị null trong cột 'quantity_sold' bằng 0
dt['quantity_sold'] = dt['quantity_sold'].fillna(0)

# Chuyển cột 'quantity_sold' sang kiểu dữ liệu int
dt['quantity_sold'] = dt['quantity_sold'].astype(int)

# Loại bỏ khoảng trắng dư thừa ở đầu của các cột 'short_description', 'product_name', 'brand_name'
dt['short_description'] = dt['short_description'].str.lstrip()
dt['product_name'] = dt['short_description'].str.lstrip()
dt['brand_name'] = dt['brand_name'].str.lstrip()

# Loại bỏ các hàng trùng lặp dựa trên cột 'id' và 'sku'
dt = dt.drop_duplicates(subset=['id'], keep=False)
dt = dt.drop_duplicates(subset=['sku'], keep=False)

# Tìm tất cả các cột có dữ liệu giống nhau
columns_with_same_values = dt.columns[dt.nunique() == 1]

# Loại bỏ các cột có dữ liệu giống nhau
dt.drop(columns=columns_with_same_values, inplace=True)

# In ra số lượng giá trị null sau khi làm sạch
print(dt.isnull().sum())

# Lưu dữ liệu đã được làm sạch vào một tệp CSV mới
dt.to_csv('clean_data/clean_data.csv', index=False, encoding='utf-8-sig')
```

### 2.3 So sánh sự thay đổi

#### So sánh sự thay đổi của các cột quan trọng

```python
import matplotlib.pyplot as plt
import pandas as pd

# Đọc dữ liệu từ tệp CSV
raw_data = pd.read_csv('raw_data/raw_data.csv', encoding='utf-8-sig')
clean_data = pd.read_csv('clean_data/clean_data.csv', encoding='utf-8-sig')

# Chọn các biến quan trọng cần so sánh
important_features = ['price', 'list_price', 'discount', 'discount_rate', 'rating_average', 'review_count', 'quantity_sold', 'brand_name']

# Vẽ đồ thị histogram cho mỗi biến quan trọng trước và sau khi xử lý
fig, axes = plt.subplots(nrows=len(important_features), ncols=2, figsize=(15, 20))

for i, feature in enumerate(important_features):
    # Đồ thị trước khi xử lý
    axes[i, 0].hist(raw_data[feature], bins=30, color='blue', alpha=0.7)
    axes[i, 0].set_title(f'Before Processing - {feature}')
    axes[i, 0].set_xlabel(feature)
    axes[i, 0].set_ylabel('Frequency')
    if feature == 'brand_name':
        axes[i, 0].tick_params(axis='x', rotation=90)
    
    # Đồ thị sau khi xử lý
    axes[i, 1].hist(clean_data[feature], bins=30, color='orange', alpha=0.7)
    axes[i, 1].set_title(f'After Processing - {feature}')
    axes[i, 1].set_xlabel(feature)
    axes[i, 1].set_ylabel('Frequency')
    if feature == 'brand_name':
        axes[i, 1].tick_params(axis='x', rotation=90)
     

plt.tight_layout()
plt.show()
```

```python
columns_of_interest = ['price', 'list_price', 'discount', 'discount_rate', 'rating_average', 'review_count', 'quantity_sold', 'brand_name']
# Số lượng cột và mẫu của dữ liệu gốc
raw_samples, raw_features = raw_data.shape

# Số lượng cột và mẫu của dữ liệu sau khi làm sạch
clean_samples, clean_features = clean_data.shape
raw_selected = raw_data[columns_of_interest]
clean_selected = clean_data[columns_of_interest]

# Tính toán giá trị trung bình và độ lệch chuẩn của các cột
raw_stats = raw_selected.describe().loc[['mean', 'std']]
clean_stats = clean_selected.describe().loc[['mean', 'std']]

# Tính độ lệch giữa dữ liệu gốc và dữ liệu sau khi làm sạch
mean_diff = clean_stats.loc['mean'] - raw_stats.loc['mean']
std_diff = clean_stats.loc['std'] - raw_stats.loc['std']

# Tạo DataFrame thể hiện sự thay đổi của các cột
comparison_df = pd.DataFrame({
    'Column': mean_diff.index,
    'Raw Mean': raw_stats.loc['mean'].values,
    'Clean Mean': clean_stats.loc['mean'].values,
    'Mean Difference': mean_diff.values,
    'Raw Std': raw_stats.loc['std'].values,
    'Clean Std': clean_stats.loc['std'].values,
    'Std Difference': std_diff.values
})

print(comparison_df)
print("\n")
print("Kết luận:")

print("Dữ liệu gốc có", raw_features, "cột và", raw_samples, "mẫu.")
print("Dữ liệu sau khi làm sạch có", clean_features, "cột và", clean_samples, "mẫu.")

# Truy cập và hiển thị con số thay đổi cụ thể
for index, row in comparison_df.iterrows():
    column_name = row['Column']
    mean_diff_rounded = round(row['Mean Difference'], 2)  # Làm tròn đến 2 chữ số thập phân
    std_diff_rounded = round(row['Std Difference'], 2)  # Làm tròn đến 2 chữ số thập phân
    
    # Thêm đơn vị nếu cần
    mean_diff_formatted = f"{mean_diff_rounded} VND" if column_name in ['price', 'discount'] else mean_diff_rounded
    std_diff_formatted = f"{std_diff_rounded} VND" if column_name in ['price', 'discount'] else std_diff_rounded
    
    print(f"Cột: {column_name}")
    print(f"Thay đổi trung bình: {mean_diff_formatted}")
    print(f"Thay đổi độ lệch chuẩn: {std_diff_formatted}")
    print("\n")

raw_brand_count = raw_data['brand_name'].nunique()
clean_brand_count = clean_data['brand_name'].nunique()

print("Số lượng thương hiệu trong dữ liệu gốc:", raw_brand_count)
print("Số lượng thương hiệu trong dữ liệu sau khi làm sạch:", clean_brand_count)

brand_change = clean_brand_count - raw_brand_count
if brand_change > 0:
    print(f"Số lượng thương hiệu đã tăng thêm {brand_change} so với dữ liệu gốc.")
elif brand_change < 0:
    print(f"Số lượng thương hiệu đã giảm đi {abs(brand_change)} so với dữ liệu gốc.")
else:
    print("Không có sự thay đổi về số lượng thương hiệu giữa dữ liệu gốc và dữ liệu sau khi làm sạch.")
```

## 4. Trực quan hoá mối quan hệ đa biến

### Price & Brand_name

```python
import pandas as pd
from sklearn.cluster import KMeans


clean_data['encoded_brand'] = encoded_brand_vector

# Chọn các đặc trưng cần thiết cho việc phân loại
features = ['price']

# Tiêu chuẩn hóa dữ liệu nếu cần thiết
scaler = StandardScaler()
data_scaled = scaler.fit_transform(clean_data[features])

# Áp dụng phương pháp K-means với số lượng nhóm mong muốn
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Gắn nhãn phân loại vào dữ liệu
clean_data['cluster'] = clusters

# Tính tổng số lượng thương hiệu trong mỗi nhóm phân loại
high_price_brands = clean_data[clean_data['cluster'] == 0]['encoded_brand'].nunique()
medium_price_brands = clean_data[clean_data['cluster'] == 1]['encoded_brand'].nunique()
low_price_brands = clean_data[clean_data['cluster'] == 2]['encoded_brand'].nunique()

# Tạo DataFrame từ dữ liệu đã tính toán
price_counts = pd.DataFrame({
    'Price Level': ['High', 'Medium', 'Low'],
    'Brand Count': [high_price_brands, medium_price_brands, low_price_brands]
})

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))
plt.bar(price_counts['Price Level'], price_counts['Brand Count'], color='skyblue')
plt.xlabel('Price Level')
plt.ylabel('Brand Count')
plt.title('Tổng số lượng thương hiệu theo nhóm giá')
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
kmeans = KMeans(n_clusters=4)
X = clean_data[['encoded_brand', 'price']]

kmeans.fit(X)
clean_data['cluster'] = kmeans.labels_
plt.figure(figsize=(12, 8))

sns.scatterplot(data=clean_data, x='brand_name', y='price', hue='cluster', palette='Set1')

plt.xlabel('Brand Name')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.title('Biểu đồ phân cụm dữ liệu')
plt.legend(title='Cluster')
plt.show()
```

### Tổng số lượng thương hiệu theo mức đánh giá

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('./clean_data/clean_data.csv')

# Chọn các đặc trưng cần thiết cho việc phân loại (ví dụ: mức đánh giá)
features = ['rating_average']

# Tiêu chuẩn hóa dữ liệu nếu cần thiết
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[features])

# Áp dụng phương pháp K-means với số lượng nhóm mong muốn
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Gắn nhãn phân loại vào dữ liệu
data['classification'] = clusters

# Tính tổng số lượng thương hiệu theo mức đánh giá trong mỗi nhóm phân loại
high_rating_count = data[data['classification'] == 0]['brand_name'].nunique()
medium_rating_count = data[data['classification'] == 1]['brand_name'].nunique()
low_rating_count = data[data['classification'] == 2]['brand_name'].nunique()

# Tạo DataFrame từ dữ liệu đã tính toán
rating_counts = pd.DataFrame({
    'Rating Level': ['High', 'Medium', 'Low'],
    'Brand Count': [high_rating_count, medium_rating_count, low_rating_count]
})

# In ra bảng thống kê
print(rating_counts)

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))
plt.bar(rating_counts['Rating Level'], rating_counts['Brand Count'], color='skyblue')
plt.xlabel('Rating Level')
plt.ylabel('Brand Count')
plt.title('Tổng số lượng thương hiệu theo mức đánh giá')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Thêm lưới ngang
plt.show()
```

```python
def elbow_method(data, features):
    # Standardize the data if necessary
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data[features])

    # Number of clusters to test (k)
    k_values = range(1, 8)

    # Distortion (inertia) for each k
    distortions = []
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data_scaled)
        distortions.append(kmeans.inertia_)

    # Plot the Elbow Method
    plt.plot(k_values, distortions, 'bx-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Distortion')
    plt.title('Elbow Method for Clustering')
    plt.show()

# Example usage:
features = ['price']
elbow_method(clean_data, features)
```
