
Dựa trên biểu đồ ma trận tương quan, chúng ta có thể đưa ra những nhận xét sau về các đặc trưng:

#### price, list_price

price, list_price có mối tương quan dương cao, điều này dễ hiểu vì giá niêm yết thường ảnh hưởng đến giá.

#### quantity_sold

quantity_sold có mối tương quan lớn với số lượng đánh giá.
